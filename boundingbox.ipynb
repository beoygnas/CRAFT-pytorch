```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import matplotlib.patches as patches

im = Image.open('./img/img6.jpg')

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect = patches.Rectangle((107, 20), 178 - 107, 77 - 20, linewidth=1, edgecolor='r', facecolor='none')
rect = patches.Rectangle((428, 27), 581 - 428, 87 - 27, linewidth=1, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from matplotlib import pyplot as plt

x, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)

plt.scatter(x[:, 0], x[:, 1], c='white', marker='o', edgecolors='black', s=50)
plt.grid()
plt.tight_layout()
plt.show()

kmeans = KMeans(n_clusters=3, init="random", n_init=10, max_iter=300, tol=1e-04, random_state=0)
pred = kmeans.fit_predict(x)
```

```python
f = open("./result/res_img8.txt", 'r')

box_x1 = []
box_x2 = []
box_y1 = []
box_y2 = []
box_width = []
box_height = []


while True:
    line = f.readline()
    if not line: break
    coord = line.strip().split(",")
    coord = [int(item) for item in coord]
    box_x1.append(coord[0])
    box_y1.append(coord[1])
    box_x2.append(coord[2])
    box_y2.append(coord[5])
    box_width.append(coord[2] - coord[0])
    box_height.append(coord[5] - coord[1])
    

oneline_x1 = -1 
oneline_y1 = -1
oneline_x2 = -1
oneline_y2 = -1

im = Image.open('./img/img8.jpg')
fig, ax = plt.subplots()
ax.imshow(im)


print(box_height)
for i in range(0, len(box_y1)) :

    if i != 0 and abs(box_y1[i] - box_y1[i-1]) >= box_height[i] :
        print(box_y1[i] - box_y1[i-1], box_height[i])
        rect = patches.Rectangle((oneline_x1, oneline_y1), oneline_x2 - oneline_x1, oneline_y2 - oneline_y1, linewidth=1, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
        oneline_x1 = -1
        oneline_x2 = -1
        oneline_y1 = -1
        oneline_y2 = -1
        

    if box_x1[i] < oneline_x1 or oneline_x1 == -1 :
        oneline_x1 = box_x1[i]
    if box_y1[i] < oneline_y1 or oneline_y1 == -1 :
        oneline_y1 = box_y1[i]
    if box_x2[i] > oneline_x2 or oneline_x2 == -1 :
        oneline_x2 = box_x2[i]
    if box_y2[i] > oneline_y2 or oneline_y2 == -1 :
        oneline_y2 = box_y2[i]
    
    
plt.savefig("./img/High resoltion.png")
plt.show()
    
    



f.close()
```
